### Dataset 1: Patient Readmission (Synthetic Data)

This section generates a synthetic dataset based on the specified requirements for a patient readmission prediction task. The dataset contains 500 entries with various features and a binary target variable indicating readmission within 30 days.

```python
import pandas as pd
import numpy as np

# Number of patients
n_patients = 500

# Generate synthetic data
data = {
    'อายุ': np.random.randint(20, 90, n_patients), # Age from 20 to 89
    'เพศ': np.random.choice(['ชาย', 'หญิง'], n_patients), # Gender
    'โรคหลัก': np.random.choice(['เบาหวาน', 'ความดันโลหิตสูง', 'โรคหัวใจ', 'หอบหืด', 'ไข้หวัดใหญ่'], n_patients), # Main Disease
    'จำนวนโรคร่วม': np.random.randint(0, 6, n_patients), # Number of Comorbidities from 0 to 5
    'วันนอนครั้งที่แล้ว': np.random.randint(1, 31, n_patients), # Previous Stay Duration from 1 to 30 days
    'กลับมาใน 30 วัน (0/1)': np.random.choice([0, 1], n_patients, p=[0.7, 0.3]) # Target: 0 for no readmission, 1 for readmission (with 30% readmission rate)
}

df_readmission = pd.DataFrame(data)

print(f"Generated dataset with {len(df_readmission)} patients.")
print("First 5 rows of the dataset:")
display(df_readmission.head())

print("\nDataset Information:")
display(df_readmission.info())

print("\nValue counts for the target variable:")
display(df_readmission['กลับมาใน 30 วัน (0/1)'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd

# 1. Separate features (X) and target (y)
X = df_readmission.drop('กลับมาใน 30 วัน (0/1)', axis=1)
y = df_readmission['กลับมาใน 30 วัน (0/1)']

# 2. Preprocess features: One-hot encode categorical variables
X_processed = pd.get_dummies(X, columns=['เพศ', 'โรคหลัก'], drop_first=True)

print("Features after One-Hot Encoding:")
display(X_processed.head())

# 3. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nTraining data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")

# 4. Initialize and train a Logistic Regression model
model = LogisticRegression(random_state=42, solver='liblinear') # 'liblinear' is good for small datasets and handles both L1/L2 penalties
model.fit(X_train, y_train)

print("\nLogistic Regression model trained successfully.")

# 5. Evaluate the model
y_pred = model.predict(X_test)

print("\nModel Evaluation:")
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

### Applying Random Forest Classifier

Given the limitations of the Logistic Regression model, particularly its inability to predict the minority class, we will now try a Random Forest Classifier. Random Forests are ensemble learning methods that can often provide better performance and handle imbalanced datasets more robustly than simpler models.

```python
from sklearn.ensemble import RandomForestClassifier

# 1. Initialize and train a Random Forest Classifier
# Using class_weight='balanced' to handle class imbalance
rf_model = RandomForestClassifier(random_state=42, class_weight='balanced')
rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")

# 2. Evaluate the Random Forest model
y_pred_rf = rf_model.predict(X_test)

print("\nRandom Forest Model Evaluation:")
print("Classification Report:")
print(classification_report(y_test, y_pred_rf))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
```

### Applying SMOTE to address class imbalance

Given that the Random Forest model still struggles with identifying the minority class, we will now apply **SMOTE (Synthetic Minority Over-sampling Technique)** to the training data. SMOTE creates synthetic samples of the minority class, helping the model learn its patterns more effectively without overfitting to existing minority samples.

```python
# Install imblearn if not already installed
!pip install imbalanced-learn

from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Apply SMOTE to the training data
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print(f"Shape of X_train before SMOTE: {X_train.shape}")
print(f"Shape of X_train after SMOTE: {X_train_res.shape}")
print(f"Value counts of y_train before SMOTE:\n{y_train.value_counts()}")
print(f"Value counts of y_train after SMOTE:\n{y_train_res.value_counts()}")
```

```python
# 1. Initialize and train a Random Forest Classifier on SMOTE-resampled data
# No need for class_weight='balanced' now as SMOTE balances the classes
rf_model_smote = RandomForestClassifier(random_state=42)
rf_model_smote.fit(X_train_res, y_train_res)

print("Random Forest model trained successfully with SMOTE data.")

# 2. Evaluate the Random Forest model with SMOTE data
y_pred_rf_smote = rf_model_smote.predict(X_test)

print("\nRandom Forest Model Evaluation (with SMOTE-resampled training data):")
print("Classification Report:")
print(classification_report(y_test, y_pred_rf_smote))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf_smote))
```
